```python
nums=covid.select_dtypes(include=['int64','float64'])
nums
```

```python

from scipy import stats
for col in nums.columns:
    plt.figure(figsize=(14,4))
    plt.subplot(121)
    sns.histplot(nums[col])
    plt.title(col)

    plt.subplot(122)
    stats.probplot(nums[col],plot = plt)
    plt.title(col)

    plt.show()
```

```python
for i in nums:
  lst1=covid[i].value_counts()
  tisland=covid[i].unique()
  plt.pie(lst1,labels=tisland,labeldistance=1)
  plt.show()
```

```python
sns.FacetGrid(covid, col = 'SEX', height = 4)
sns.histplot(covid['CLASIFFICATION_FINAL'])
```

```python
for i in nums:
  sns.boxplot(x=nums[i])
  print("boxplot for :"+i)
  plt.show()
```

```python
Q1=nums["AGE"].quantile(0.25)
Q3=nums["AGE"].quantile(0.75)
IQR=Q3 - Q1
lower_bound=Q1 - 1.5 * IQR

upper_bound=Q3 + 1.5 * IQR


new=[]
for i in nums["AGE"]:
    if  (i>upper_bound):
           new.append(upper_bound)
    elif (i<lower_bound):
        new.append(lower_bound)
    else:
         new.append(i)

nums["AGE"]=new
```

```python
plt.figure(figsize=(15,15))
corr = nums.corr()
sns.heatmap(corr, annot=True)
plt.title("Features Correlation", fontsize = 15)
plt.show()
```
